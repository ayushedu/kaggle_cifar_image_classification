#Mahotas demo for Image classification

```python
import mahotas as mh
import numpy as np
```

##Load image

```python
image_color = mh.imread('data/train/train/1.png')
image_color.shape
```

##Display image

```python
# Show image
from matplotlib import pyplot as plt
%matplotlib inline
plt.imshow(image_color)
plt.show()
```

##Convert image to grayscale

```python
image = mh.colors.rgb2gray(image_color,dtype=np.uint8)
print 'Image shape', image.shape
%matplotlib inline
plt.imshow(image)
plt.gray()
```

##Threshold the image

```python
thresh = mh.thresholding.otsu(image)
print('Otsu threshold is {}.'.format(thresh))
plt.imshow(image > thresh)
```

```python
features = mh.features.haralick(image)
features_mean = np.mean(features, axis=0)
features_all = np.ravel(features)
```
